### Random Forest

```python
# Divide dataset into X and y
features = ['economy_fiscal_news_sentiment_score',
       'economy_fiscal_news_sentiment_label',
       'economy_macro_news_sentiment_score',
       'economy_macro_news_sentiment_label',
       'economy_monetary_news_sentiment_score',
       'economy_monetary_news_sentiment_label',
       'financial_market_news_sentiment_score',
       'financial_market_news_sentiment_label', 
       'ticker_news_compound',
       'volume', 
       'MACD_signal',
       'OBV ema',
       'RSI']

y = data['adj_close']
X = data[features]

# Scale the data
scaler = StandardScaler()
X = scaler.fit_transform(X)

# Divide data into test and train test
X_train = X[:120, :]
X_test = X[120:, :]

y_train = y[:120]
y_test = y[120:]
```

```python
forest = RandomForestRegressor(n_jobs=-1)
forest.fit(X_train, y_train)
```

```python
feature_importance = forest.feature_importances_
```

```python
plt.bar(height=feature_importance, x=np.arange(0, len(feature_importance)))
```
